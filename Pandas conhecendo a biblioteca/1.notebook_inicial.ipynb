## Qual o valor médio de aluguel por tipo de imóvel?

```python
#Algumas perguntas que podemos fazer nesse momento:

#Quais os valores médios de aluguel por tipo de imóvel?

#Qual o percentual de cada tipo de imóvel na nossa base de dados?
```

```python
dados.head()
```

```python
#mean = metodo para calcular a média
dados['Valor'].mean()
```

```python
#agrupar dados com cada tipo de imóvel = groupby
dados.groupby('Tipo').mean(numeric_only=True)
```

```python
#agrupar dados com cada tipo de imóvel = groupby
dados.groupby('Tipo')['Valor'].mean()
```

```python
#melhorar a visualização
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
#plotar gráfico utilizando o dataframe, defina uma variável para o dataframe e depois utilize o método plot,
# defina o tipo de gráfico e o título do gráfico
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color = 'purple')
#kind = tipo de gráfico
#figsize = tamanho do gráfico
#color = cor do gráfico
```

## Removendo os imóveis comerciais

```python
#REMOVER OS IMOVEIS COMERCIAIS DA BASE DE DADOS POIS PODEM ALGUNS PODEM INFLUENCIAR NOS RESULTADOS,
# POIS O VALOR DE ALUGUEL DE UM IMOVEL COMERCIAL É MUITO MAIOR QUE UM RESIDENCIAL

dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
'Prédio Inteiro', 'Loja/Salão',
'Galpão/Depósito/Armazém',
'Casa Comercial', 'Terreno Padrão',
'Loja Shopping/ Ct Comercial',
'Box/Garagem', 'Chácara',
'Loteamento/Condomínio', 'Sítio',
'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
dados.query('@imoveis_comerciais in Tipo')
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df.head()
```

```python
df.Tipo.unique()
```

```python
# 1.VALOR MÉDIO DE ALUGUEL POR TIPO DE IMÓVEL (SEM IMÓVEIS COMERCIAIS)
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color = 'purple')
```
